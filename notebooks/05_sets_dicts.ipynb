# Sets et Dictionnaires

Les structures de donnees les plus utiles en Python. Si tu maitrises ca, t'as fait 80% du chemin.

## Sets - Elements uniques

```python
# Creation
s = {1, 2, 3}
s = set([1, 2, 2, 3, 3, 3])  # Doublons vires
print(s)
```

```python
# Operations
s.add(4)
s.remove(1)       # Erreur si absent
s.discard(10)     # Pas d'erreur si absent

# Test d'appartenance O(1)
print(2 in s)
```

```python
# Operations ensemblistes
a = {1, 2, 3}
b = {2, 3, 4}

print(f"Union:        {a | b}")
print(f"Intersection: {a & b}")
print(f"Difference:   {a - b}")
```

## Pattern: Positions visitees

```python
# Super courant en AoC
visited = set()
pos = (0, 0)

moves = [(1, 0), (0, 1), (1, 0), (-1, 0)]

for dx, dy in moves:
    pos = (pos[0] + dx, pos[1] + dy)
    
    if pos in visited:
        print(f"Deja visite: {pos}")
    else:
        visited.add(pos)
        
print(f"Positions uniques: {len(visited)}")
```

## Dictionnaires

```python
# Creation
scores = {"Alice": 100, "Bob": 85}

# Acces
print(scores["Alice"])
print(scores.get("Eve", 0))  # Defaut si absent

# Modification
scores["Alice"] = 110
scores["Eve"] = 95
```

```python
# Iteration
for nom, score in scores.items():
    print(f"{nom}: {score}")
```

## defaultdict - Plus de KeyError

```python
from collections import defaultdict

# Comptage
compteur = defaultdict(int)
for lettre in "abracadabra":
    compteur[lettre] += 1  # Pas besoin d'initialiser!

print(dict(compteur))
```

```python
# Graphe
graphe = defaultdict(list)
graphe["A"].append("B")
graphe["A"].append("C")
graphe["B"].append("D")

print(dict(graphe))
```

## Counter - Comptage automatique

```python
from collections import Counter

texte = "abracadabra"
freq = Counter(texte)

print(freq)
print(f"Les 3 plus frequents: {freq.most_common(3)}")
```
